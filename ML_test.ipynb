```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

df=pd.read_csv('대덕구.csv', header = None)
# 대덕구.csv 유성구.csv 서구.csv
```

```python
data1= df.loc[:,0:3]
data2 = df.loc[:,7]
data3 = data1.assign(BAL=data2)

data = data3
target = df.loc[:,5]
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.3, shuffle=False)
```

```python
import matplotlib.pyplot as plt

train_acc = []
test_acc = []

for n in range(1,10):
    classifier = KNeighborsClassifier(n_jobs=-1, n_neighbors=n)
    classifier.fit(x_train, y_train)
    prediction = classifier.predict(x_test)
    train_acc.append(classifier.score(x_train, y_train))
    test_acc.append((prediction==y_test).mean())
```

```python
plt.figure(figsize=(12, 9))
plt.plot(range(1, 10), train_acc, label='TRAIN set')
plt.plot(range(1, 10), test_acc, label='TEST set')
plt.xlabel("neighbors k")
plt.ylabel("accuracy")
plt.xticks(np.arange(1, 10, step=1))
plt.legend()
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(x_train, y_train)
prediction = classifier.predict(x_test)

print('대덕구')
classifier.score(x_test, y_test)
```
