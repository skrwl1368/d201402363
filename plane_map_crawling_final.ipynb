```python
bb = data[0]
cc = data[1]
aa = data[2]
dd = data[3]
ee = data[4]
ff = data[5]
main_area = []
bedroom = []
bath = []
con_rate = []
porch_structure = []
administrative_expenses = []
img_url = []

for i in range(len(ff)):
    img_url.append(ff.values[i])
for i in range(len(bb)):
    bb.values[i] = str(bb.values[i]).rstrip('㎡')
    bb.values[i]= math.trunc(float(bb.values[i]))
    main_area.append(bb.values[i])

for i in range(len(cc)):
    con_rate.append(float(cc.values[i].rstrip('%')))

for i in range(len(ee)):
    if ee.values[i] == '없음':
        administrative_expenses .append(float(0))
#         os.remove(ff.values[i])
        continue
    ee.values[i] = ee.values[i].rstrip('원')
    ee.values[i] = ee.values[i].replace(',', '')
    administrative_expenses .append(float(ee.values[i]))
    

for i in range(len(aa)):
    bedroom.append(float(aa.values[i][0]))
    bath.append(float(aa.values[i][5]))


for i in range(len(aa)):
    bedroom.append(float(aa.values[i][0]))
    bath.append(float(aa.values[i][5]))

for i in range(len(dd)):
    if dd.values[i] == '복도식':
        porch_structure.append(int(0))
    elif dd.values[i] == '계단식':
        porch_structure.append(int(1))
    elif dd.values[i] == '복합식':
        porch_structure.append(int(2))
```

```python
data2 = pd.DataFrame()

data2.loc[:, "con_rate"] = pd.Series(con_rate)
data2.loc[:, "bedroom"] = pd.Series(bedroom)
data2.loc[:, "bath"] = pd.Series(bath)
data2.loc[:, "porch_structure"] = pd.Series(porch_structure)
data2.loc[:, "administrative_expenses"] = pd.Series(administrative_expenses)
data2.loc[:, "main_area"] = pd.Series(main_area)
data2.loc[:, "img_url"] = pd.Series(img_url)

non_expense = data2[data2["administrative_expenses"] == 0.0].index
data2 = data2.drop(non_expense)
```
